```python
import numpy as np
import matplotlib.pyplot as plt

# Функция для генерации данных
def generate_data(N):
    x_values = np.arange(1, N + 1)
    true_line = 0.5 * x_values
    noise = np.random.normal(0, 1, N)  # Генерация случайных значений с нормальным распределением
    y_values = true_line + noise
    return x_values, y_values

# Функция для выполнения линейной регрессии
def linear_regression(x_values, y_values):
    A = np.vstack([x_values, np.ones(len(x_values))]).T
    a_star, b_star = np.linalg.lstsq(A, y_values, rcond=None)[0]
    return a_star, b_star

# Функция для построения графиков
def plot_results(x_values, y_values, true_line, a_star, b_star):
    plt.scatter(x_values, y_values, label='Экспериментальные точки')
    plt.plot(x_values, true_line, label='Истинная прямая', color='green')
    plt.plot(x_values, a_star * x_values + b_star, label='Линейная регрессия', linestyle='dashed', color='red')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.title('Линейная регрессия')
    plt.show()

# Генерация данных и выполнение линейной регрессии
N = 10
x_values, y_values = generate_data(N)
a_star, b_star = linear_regression(x_values, y_values)

# Построение графика
plot_results(x_values, y_values, 0.5 * x_values, a_star, b_star)

# Вывод коэффициентов
print(f'a*: {a_star}')
print(f'b*: {b_star}')
```
